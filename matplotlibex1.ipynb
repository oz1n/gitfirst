```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False #用来正常显示负号
```

##### 創建一個折線圖：

```python
plt.figure()  #創建一個畫布
```

```python
x = np.array([2, 3, 5, 4]) #給定座標
y = np.array([3, 4, 6, 4])
z = 2 * x + 0.5
```

##### 在繪圖之前可以添加一些表的信息：

```python
plt.title("第一條折線圖")  # 給圖片一個標題，添加中文标题乱码
plt.xticks(x)  # X軸的刻度，傳X表示根據X的值顯示刻度
plt.yticks(np.arange(15))
plt.xlabel("X 軸")  #X軸顯示的標籤
plt.ylabel("Y 軸")  #Y軸顯示的標籤
# 正式绘图：
plt.plot(x, y, linestyle = "-.", linewidth = 2.5, color = "hotpink", marker = "o",label = 'Lines') #創建折線圖，並給參數
plt.plot(x, z, linestyle = ":", linewidth = 2.5, color = "blue", marker = "d",label = 'X + 1') #創建第二條折線圖
plt.legend()  #顯示圖例
```

##### 散點圖：

```python
plt.figure()
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
```

```python
plt.scatter(x, y,) #創建散點圖
plt.show()
```

##### 子圖：

```python
p = plt.figure()
p.add_subplot(2,1,1) #創建一個子圖
plt.scatter(x, y)
p.add_subplot(2,1,2) #創建第二個子圖
plt.plot(x, y)
plt.legend(["Sin圖"])
plt.show()
```

##### 直方圖：

```python
plt.figure()
salary = np.array([1050, 2000, 3050, 2800, 4050, 4000, 6500, 8000, 5500, 6000, 7000, 8500, 8800, 3600, 1200, 2500,4200])
group = np.array([1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000])
plt.hist(salary, group, color = "lightpink")
plt.show()
```

##### 柱狀圖：

```python
plt.figure()
x = np.array([1,2,3])
y = np.arange(1000, 4000,1000)
plt.xlabel("數量")
plt.ylabel("金額")
plt.bar(x, y, color = ['hotpink', 'skyblue', 'palegreen'])
for x, y in zip(x, y):
    plt.text(x, y + 20, y,ha = "center", fontsize = 11)

plt.show()
```

##### 通過直方圖了解<font color= "hotpink"> 正態分佈 </font>和 <font color= "skyblue"> 平均分佈 </font>

```python
a = np.random.randn(10000)
plt.figure()
plt.hist(a, 20, color = "darkviolet") #圖片顯示靠近 0 的地方數據最多，離 0 越遠，數據越少，符合正態分佈的特點。
plt.show()
```

```python
b = np.random.rand(10000)
plt.figure()
plt.hist(b, 20, color = "turquoise") # 圖片顯示每個數據的數量基本相同，符合平均分佈的特徵。
plt.show()
```

##### 餅圖：

```python
data = np.array([100, 200, 300])
explode = [0.01, 0.01, 0.1] #設置圓心距，即是設置突出的部分，0.1就是突出的部分
labels = ['5年','10年','15年']

plt.figure()
plt.pie(data, explode = explode, labels = labels,     #繪製餅圖，
       colors = ["hotpink", "skyblue", "palegreen"],  #顯示顏色
       autopct = "%.1f%%", labeldistance = 1.2,       #autopct表示顯示百分比的格式, labeldistance 表示標籤與餅圖的距離
       pctdistance = 0.5)                             #pctdistance 表示百分比的位置，數字表示距離一個半徑的百分比
plt.show()
```
