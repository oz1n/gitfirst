```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
y = [200,400,300]
x = range(1,4)
sns.set(style = "darkgrid", context = "notebook", font = "SimHei", font_scale = 1.1) #font 添加中文支持
plt.figure()
plt.bar(x, y, tick_label = x, color = ["orange","hotpink","skyblue"])
for x,y in zip(x, y):
    plt.text(x, y+0.1, y, ha = "center", fontsize = 10)
plt.title("seaborn 中文測試")
plt.show()
```
