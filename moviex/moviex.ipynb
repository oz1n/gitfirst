```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies.info()
```

#### 合併表格： 利用merge來通過ID 合併，但兩張表格的ID不同，需要先改名字。

```python
credits.rename(columns = {'movie_id' : 'id'}, inplace = True) #注意如果不用inplace原表格不會改變，這樣進行後續操作都會有誤。
credits.info()
```

```python
tmdb = pd.merge(credits, movies, on = ["title", "id"])
tmdb.info()
```

##### 電影類型如何隨著時間的推移發生變化？

```python
tmdb.loc[tmdb.release_date.isnull(), 'title'] #找到電影中日期為空值的行。
```

```python
tmdb.release_date.fillna('2014-06-01', inplace = True)
tmdb.release_date[4553]
```

```python
tmdb['genres'] = tmdb['genres'].transform(json.loads) 
```

```python
type_set = set()
def get_movie_type(element):
    type_list = []
    for value in element:
        name = value['name']
        type_list.append(name)
        type_set.add(name)
    return ' '.join(type_list)
```

```python
tmdb['genres'] = tmdb['genres'].transform(get_movie_type)
```

```python
tmdb['release_year'] = pd.to_datetime(tmdb['release_date']).dt.year
```

```python
for movie_type in list(type_set):
    tmdb[movie_type] = tmdb['genres'].str.contains(movie_type).transform(lambda x:1 if x else 0)
```

```python
tmdb.shape
```

```python
groupby_year = tmdb.groupby('release_year')[list(type_set)].sum()
groupby_year
```

```python
plt.figure(figsize= (16,8))
x = groupby_year.index
for movie_type in list(type_set):
    y = groupby_year[movie_type]
    plt.plot(x, y)
plt.legend(list(type_set))
plt.show()
```
