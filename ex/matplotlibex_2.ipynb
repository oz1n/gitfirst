##### 畫一個 2000 年到 2017 年第一季度的第一產業對比圖

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('国民经济核算季度数据.npz', allow_pickle = True)
```

```python
columns = data["columns"]
values = data["values"]
```

```python
first_q_data = values[::4,3]  #篩選產值數據
xlabels = values[::4,1]       #篩選日期
#shortlabel = []           #節選日期標籤
#for xl in xlabels:       #讓X軸的日期只顯示年份
#    shortlabel.append(xl[0:4])

shortlabel = np.array([])
for x in xlabels:
    shortlabel = np.append(shortlabel, x[0:4])

plt.figure(figsize = (12,6))  #設定畫布的大小
plt.bar(shortlabel, first_q_data, color = "peachpuff") #繪出柱狀圖數據
for shortlabel, first_q_data in zip(shortlabel, first_q_data):   #給每一個柱狀圖商添加數據文本
    plt.text(shortlabel, first_q_data + 50, first_q_data, ha = "center", fontsize = 9) 
plt.title("2000年至2017年第一季度第一產業增加值")
plt.xlabel("年份")
plt.ylabel("產值")
plt.show()
```
